### Towards feature extraction

```python
# missing values
paid.isnull().sum()
```

```python
# amount of money per each customer

avg_pay = paid.groupby("customer_id", as_index=False)["total_price"].mean()

avg_pay.sort_values(["total_price"], ascending=False)
```

```python
# total price per customer mean 
avg_pay ['total_price'].mean()
```

```python
# number of orders per customer
num_orders_customer = paid.groupby("customer_id", as_index=False)["id"].count()
num_orders_customer.sort_values(["id"], ascending=False)

num_orders_customer=num_orders_customer.rename(columns={'id': 'num_orders'})
```

```python
# join two features

join_features = pd.concat([avg_pay, num_orders_customer], axis=1)
join_features.head()
```
